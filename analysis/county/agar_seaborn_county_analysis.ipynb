```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

sns.set(font_scale=1.5)

%matplotlib inline
```

```python
# read NEX
nex = pd.read_csv("nex_gmfd_stats.csv")
nex["GEOID"] = nex["GEOID"].astype(str).str.zfill(5)
nex.set_index(["AgVar","GEOID"], inplace = True)
```

```python
nex.head()
```

```python
# read CMIP
cmip = pd.read_csv("cmip_gmfd_stats.csv")
cmip["GEOID"] = cmip["GEOID"].astype(str).str.zfill(5)
cmip.set_index(["AgVar","GEOID"], inplace = True)
```

```python
cmip.head()
```

```python
# Plot GDD improvement
names = [name.replace("_RMSE","") for name in cmip.columns[::2].values]
cmip_nat_gdd = cmip.groupby("AgVar").mean().loc["gdd"]
nex_nat_gdd = nex.groupby("AgVar").mean().loc["gdd"]

fig, ax = plt.subplots(1,1, figsize = (10,10))
plt.scatter(cmip_nat_gdd[:-2:2],cmip_nat_gdd[1:-2:2],c = "C3",s=80,zorder=2)
plt.scatter(nex_nat_gdd[:-2:2],nex_nat_gdd[1:-2:2],c = "C2",s=80,zorder=2)

plt.scatter(cmip_nat_gdd[-2],cmip_nat_gdd[-1],c = "C3",s=80,marker="s",zorder=2)
plt.scatter(nex_nat_gdd[-2],nex_nat_gdd[-1],c = "C2",s=80,marker="s",zorder=2)
for i in range(int(len(cmip_nat_gdd)/2)):
    ax.arrow(cmip_nat_gdd[::2][i], cmip_nat_gdd[1::2][i], -cmip_nat_gdd[::2][i]+nex_nat_gdd[::2][i], -cmip_nat_gdd[1::2][i]+nex_nat_gdd[1::2][i],color="grey",zorder=1,alpha=0.6)
ax.set_xlabel("RMSE")
ax.set_ylabel("STD ratio (days)")
plt.show()
```

```python
# Plot eGDD improvement
cmip_nat_egdd = cmip.groupby("AgVar").mean().loc["egdd"]
nex_nat_egdd = nex.groupby("AgVar").mean().loc["egdd"]

fig, ax = plt.subplots(1,1, figsize = (10,10))
plt.scatter(cmip_nat_egdd[:-2:2],cmip_nat_egdd[1:-2:2],c = "C3",s=80,zorder=2)
plt.scatter(nex_nat_egdd[:-2:2],nex_nat_egdd[1:-2:2],c = "C2",s=80,zorder=2)

plt.scatter(cmip_nat_egdd[-2],cmip_nat_egdd[-1],c = "C3",s=80,marker="s",zorder=2)
plt.scatter(nex_nat_egdd[-2],nex_nat_egdd[-1],c = "C2",s=80,marker="s",zorder=2)

for i in range(int(len(cmip_nat_egdd)/2)):
    ax.arrow(cmip_nat_egdd[::2][i], cmip_nat_egdd[1::2][i], -cmip_nat_egdd[::2][i]+nex_nat_egdd[::2][i], -cmip_nat_egdd[1::2][i]+nex_nat_egdd[1::2][i],color="grey",zorder=1,alpha=0.6)
ax.set_xlabel("RMSE")
ax.set_ylabel("STD ratio (days)")
plt.show()
```

```python
# Plot precip improvement
cmip_nat_prcp = cmip.groupby("AgVar").mean().loc["prcp"]
nex_nat_prcp = nex.groupby("AgVar").mean().loc["prcp"]

fig, ax = plt.subplots(1,1, figsize = (10,10))
plt.scatter(cmip_nat_prcp[:-2:2],cmip_nat_prcp[1:-2:2],c = "C3",s=80,zorder=2)
plt.scatter(nex_nat_prcp[:-2:2],nex_nat_prcp[1:-2:2],c = "C2",s=80,zorder=2)

plt.scatter(cmip_nat_prcp[-2],cmip_nat_prcp[-1],c = "C3",s=80,marker="s",zorder=2)
plt.scatter(nex_nat_prcp[-2],nex_nat_prcp[-1],c = "C2",s=80,marker="s",zorder=2)
for i in range(int(len(cmip_nat_prcp)/2)):
    ax.arrow(cmip_nat_prcp[::2][i], cmip_nat_prcp[1::2][i], (-cmip_nat_prcp[::2]+nex_nat_prcp[::2])[i], (-cmip_nat_prcp[1::2]+nex_nat_prcp[1::2])[i],color="grey",zorder=1,alpha=0.6,head_width=False,width=0.0001)
ax.set_xlabel("RMSE")
ax.set_xlabel("RMSE (m m$^{-2}$)")
plt.show()
```
