```python
import numpy as np
import pandas as pd
import geopandas as gp
from shapely.geometry import Point
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib import gridspec
plt.rcParams.update({'font.size': 18})

cm_data = np.loadtxt("/Users/davidlafferty/Projects/misc/ScientificColourMaps5/batlow/batlow.txt")[::-1]
batlow_map = LinearSegmentedColormap.from_list("Batlow", cm_data)

cm_data = np.loadtxt("/Users/davidlafferty/Projects/misc/ScientificColourMaps5/cork/cork.txt")
cork_map = LinearSegmentedColormap.from_list("Cork", cm_data)

cm_data = np.loadtxt("/Users/davidlafferty/Projects/misc/ScientificColourMaps5/berlin/berlin.txt")
berlin_map = LinearSegmentedColormap.from_list("Berlin", cm_data)

cm_data = np.loadtxt("/Users/davidlafferty/Projects/misc/ScientificColourMaps5/lapaz/lapaz.txt")
lapaz_map = LinearSegmentedColormap.from_list("LaPaz", cm_data)

cm_data = np.loadtxt("/Users/davidlafferty/Projects/misc/ScientificColourMaps5/tokyo/tokyo.txt")
tokyo_map = LinearSegmentedColormap.from_list("Tokyo", cm_data)

cm_data = np.loadtxt("/Users/davidlafferty/Projects/misc/ScientificColourMaps5/bamako/bamako.txt")
bamako_map = LinearSegmentedColormap.from_list("Bamako", cm_data)

cm_data = np.loadtxt("/Users/davidlafferty/Projects/misc/ScientificColourMaps5/turku/turku.txt")[::-1]
turku_map = LinearSegmentedColormap.from_list("Turku", cm_data)

# import seaborn as sns; sns.set_style("whitegrid");
```

```python
# Import plotting shapefiles
county_shp = gp.read_file('../data/conus_shp/conus_plot/final/counties_contig_plot.shp')
county_shp["GEOID"] = county_shp["GEOID"].astype(str).str.zfill(5)
county_shp['area'] = county_shp['geometry'].apply(lambda x: x.area)
county_shp = county_shp.query('area > 10e3')

states = gp.read_file('../data/conus_shp/conus_plot/final/states_contig_plot.shp')
coast = gp.read_file('../data/conus_shp/conus_plot/final/coast_contig_plot.shp')
```

# Plot 1) Yield Model Verification

```python
# Get results
yield_res = pd.read_csv('../data/GMFD_all/fitting/gmfd_regression_res_60-05.csv')
yield_res["GEOID"] = yield_res["GEOID"].astype(str).str.zfill(5)
yield_res['state'] = yield_res['GEOID'].apply(lambda x: x[:2])
yield_res.set_index(['GEOID'], inplace = True)
```

```python
# Plot all R2
fig, ax = plt.subplots(1,1, figsize=(18, 10))
data = yield_res.reset_index()
data_shp = pd.merge(county_shp, data, on="GEOID", how="outer", copy=False)

divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)
cax.tick_params(labelsize=24)
cax.set_label('R2')
data_shp.dropna().plot(column='R2', ax=ax, cax = cax, legend=True, legend_kwds={'orientation' : 'vertical', 'ticks' : np.arange(0.,1.1,0.2)}, vmin = 0., vmax = 1.0, cmap = turku_map)
# data_shp[data_shp[model].isna()].plot(ax=ax, color="lightgray")
states.geometry.boundary.plot(ax=ax, linewidth=0.5, edgecolor="black", color=None)
coast.geometry.boundary.plot(ax=ax, linewidth=0.5, edgecolor="black", color=None)
ax.tick_params(labelbottom=False, labelleft=False, color = 'white')

plt.suptitle('a) Yield model driven with observational data; $R^2$ (1960-2016)', y = 0.95)
plt.tight_layout()
plt.show()
```

```python
# Plot all R2 where p > 0.05
fig, ax = plt.subplots(1,1, figsize=(18, 10))
data = yield_res.query('pval < 0.05').reset_index()
data_shp = pd.merge(county_shp, data, on="GEOID", how="outer", copy=False)

divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)
cax.tick_params(labelsize=24)
cax.set_label('R2')
data_shp.dropna().plot(column='R2', ax=ax, cax = cax, legend=True, legend_kwds={'orientation' : 'vertical', 'ticks' : np.arange(0.,1.1,0.2)}, vmin = 0., vmax = 1.0, cmap = turku_map)
# data_shp[data_shp[model].isna()].plot(ax=ax, color="lightgray")
states.geometry.boundary.plot(ax=ax, linewidth=0.5, edgecolor="black", color=None)
coast.geometry.boundary.plot(ax=ax, linewidth=0.5, edgecolor="black", color=None)
ax.tick_params(labelbottom=False, labelleft=False, color = 'white')

plt.suptitle('b) Part (a) counties with F-test $p<0.05$; $R^2$ (1960-2016)', y = 0.95)
plt.tight_layout()
plt.show()
```

```python
# Store counties
p05_counties = yield_res.query('pval < 0.05').index.unique(level = 'GEOID')
```

# Plot 2) Climate models hindcasts: Yield PDFs compared to GMFD-driven yields

```python
nexIQR = pd.read_csv('./county/tail_pvals/nex_iqr_pval_60-05.csv')
nexIQR['GEOID'] = nexIQR["GEOID"].astype(str).str.zfill(5)
cmipIQR = pd.read_csv('./county/tail_pvals/cmip_iqr_pval_60-05.csv')
cmipIQR['GEOID'] = cmipIQR["GEOID"].astype(str).str.zfill(5)

nex80QR = pd.read_csv('./county/tail_pvals/nex_80qr_pval_60-05.csv')
nex80QR['GEOID'] = nex80QR["GEOID"].astype(str).str.zfill(5)
cmip80QR = pd.read_csv('./county/tail_pvals/cmip_80qr_pval_60-05.csv')
cmip80QR['GEOID'] = cmip80QR["GEOID"].astype(str).str.zfill(5)

nexSD = pd.read_csv('./county/tail_pvals/nex_sd_pval_60-05.csv')
nexSD['GEOID'] = nexSD["GEOID"].astype(str).str.zfill(5)
cmipSD = pd.read_csv('./county/tail_pvals/cmip_sd_pval_60-05.csv')
cmipSD['GEOID'] = cmipSD["GEOID"].astype(str).str.zfill(5)

nexMIN = pd.read_csv('./county/tail_pvals/nex_min_pval_60-05.csv')
nexMIN['GEOID'] = nexMIN["GEOID"].astype(str).str.zfill(5)
cmipMIN = pd.read_csv('./county/tail_pvals/cmip_min_pval_60-05.csv')
cmipMIN['GEOID'] = cmipMIN["GEOID"].astype(str).str.zfill(5)

nexMAD = pd.read_csv('./county/tail_pvals/nex_mad_pval_60-05.csv')
nexMAD['GEOID'] = nexMAD["GEOID"].astype(str).str.zfill(5)
cmipMAD = pd.read_csv('./county/tail_pvals/cmip_mad_pval_60-05.csv')
cmipMAD['GEOID'] = cmipMAD["GEOID"].astype(str).str.zfill(5)
```

```python
# Add State filter for plot
StateDrop = ['53', '41', '16', '06', '04', '49', '56', '30', '08', '35'] # 10 westernmost CONUS states
StateKeep = states['STATEFP'].unique()
locs = [np.where(StateKeep == drop) for drop in StateDrop]
StateKeep = np.delete(StateKeep, locs)
```

```python
# Filter
county_shp_plot = county_shp.copy()
county_shp_plot['state'] = county_shp_plot['GEOID'].str[:2]
county_shp_plot.set_index(['state'], inplace=True)
county_shp_plot = county_shp_plot.loc[StateKeep]
```

```python
# Plot NEX SD
model = 'gmfd_sd'
vmin = 0.
vmax = 1.

data = nexSD.copy()
data.set_index('GEOID', inplace=True)
data = data.loc[p05_counties]
data['gmfd_sd'] = 1.-data['gmfd_sd']

sig05_plot = pd.merge(county_shp_plot, data.query('gmfd_sd <= 0.05'), on="GEOID", how='inner', copy=False)
sig05_plot['geometry'] = sig05_plot['geometry'].apply(lambda x: x.centroid)

fig, ax = plt.subplots(1, 1, figsize=(18, 10))
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)
cax.tick_params(labelsize=20)
data_shp = pd.merge(county_shp_plot, data.filter([model, 'GEOID']), on="GEOID", how="inner", copy=False)
data_shp.dropna().plot(column=model, ax=ax, cax=cax, legend=True, cmap=batlow_map, vmin=vmin, vmax=vmax)
sig05_plot.plot(color = 'black', markersize=1, marker = 'o', ax = ax)
states.geometry.boundary.plot(ax=ax, linewidth=0.5, edgecolor="black", color=None)
coast.geometry.boundary.plot(ax=ax, linewidth=0.5, edgecolor="black", color=None)
ax.tick_params(labelbottom=False, labelleft=False, color = 'white')
ax.set_xlim([-500000,2600000])
# ax.set_title(title, fontsize=20)
plt.tight_layout()
# plt.savefig('./test.png', dpi=500)
plt.show()
```

```python
# Plot MAD
model = 'gmfd_mad'
vmin = 0.
vmax = 1.

data = nexMAD.copy()
data.set_index('GEOID', inplace=True)
data = data.loc[p05_counties]
data['gmfd_mad'] = 1.-data['gmfd_mad']

sig05_plot = pd.merge(county_shp_plot, data.query('gmfd_mad <= 0.05'), on="GEOID", how='inner', copy=False)
sig05_plot['geometry'] = sig05_plot['geometry'].apply(lambda x: x.centroid)

fig, ax = plt.subplots(1, 1, figsize=(18, 10))
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)
cax.tick_params(labelsize=20)
data_shp = pd.merge(county_shp_plot, data.filter([model, 'GEOID']), on="GEOID", how="inner", copy=False)
data_shp.dropna().plot(column=model, ax=ax, cax=cax, legend=True, cmap=batlow_map, vmin=vmin, vmax=vmax)
sig05_plot.plot(color = 'black', markersize=1, marker = 'o', ax = ax)
states.geometry.boundary.plot(ax=ax, linewidth=0.5, edgecolor="black", color=None)
coast.geometry.boundary.plot(ax=ax, linewidth=0.5, edgecolor="black", color=None)
ax.tick_params(labelbottom=False, labelleft=False, color = 'white')
ax.set_xlim([-500000,2600000])
# ax.set_title(title, fontsize=20)
plt.tight_layout()
# plt.savefig('./test.png', dpi=500)
plt.show()
```

```python
# Plot MAD
model = 'gmfd_min'
vmin = 0.
vmax = 1.

data = nexMIN.copy()
data.set_index('GEOID', inplace=True)
data = data.loc[p05_counties]
data[model] = 1.-data[model]

sig05_plot = pd.merge(county_shp_plot, data.query('gmfd_min >= 0.95'), on="GEOID", how='inner', copy=False)
sig05_plot['geometry'] = sig05_plot['geometry'].apply(lambda x: x.centroid)

fig, ax = plt.subplots(1, 1, figsize=(18, 10))
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)
cax.tick_params(labelsize=20)
data_shp = pd.merge(county_shp_plot, data.filter([model, 'GEOID']), on="GEOID", how="inner", copy=False)
data_shp.dropna().plot(column=model, ax=ax, cax=cax, legend=True, cmap=batlow_map, vmin=vmin, vmax=vmax)
sig05_plot.plot(color = 'white', markersize=1, marker = 'o', ax = ax)
states.geometry.boundary.plot(ax=ax, linewidth=0.5, edgecolor="black", color=None)
coast.geometry.boundary.plot(ax=ax, linewidth=0.5, edgecolor="black", color=None)
ax.tick_params(labelbottom=False, labelleft=False, color = 'white')
ax.set_xlim([-500000,2600000])
# ax.set_title(title, fontsize=20)
plt.tight_layout()
# plt.savefig('./test.png', dpi=500)
plt.show()
```
