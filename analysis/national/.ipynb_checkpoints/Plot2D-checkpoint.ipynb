```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

sns.set(font_scale=1.5)

%matplotlib inline
```

## 2D SD-RMSE plots

```python
# Read NEX
nex_rmse = pd.read_csv("../county/output/old/nex_agvar_rmse_56-05.csv")
nex_rmse["GEOID"] = nex_rmse["GEOID"].astype(str).str.zfill(5)
nex_rmse.set_index(["AgVar","GEOID"], inplace = True)

nex_sdr = pd.read_csv("../county/output/old/nex_agvar_sdratio_56-05.csv")
nex_sdr["GEOID"] = nex_sdr["GEOID"].astype(str).str.zfill(5)
nex_sdr.set_index(["AgVar","GEOID"], inplace = True)
```

```python
# Read CMIP
cmip_rmse = pd.read_csv("../county/output/old/cmip_agvar_rmse_56-05.csv")
cmip_rmse["GEOID"] = cmip_rmse["GEOID"].astype(str).str.zfill(5)
cmip_rmse.set_index(["AgVar","GEOID"], inplace = True)

cmip_sdr = pd.read_csv("../county/output/old/cmip_agvar_sdratio_56-05.csv")
cmip_sdr["GEOID"] = cmip_sdr["GEOID"].astype(str).str.zfill(5)
cmip_sdr.set_index(["AgVar","GEOID"], inplace = True)
```

```python
# Plot improvement
geoid = '17019'
agvar = 'gdd'
cmip_nat_gdd_rmse = cmip_rmse.loc[agvar, geoid]
nex_nat_gdd_rmse = nex_rmse.loc[agvar, geoid]

cmip_nat_gdd_sdr = cmip_sdr.loc[agvar, geoid]
nex_nat_gdd_sdr = nex_sdr.loc[agvar, geoid]

fig, ax = plt.subplots(1,1, figsize = (10,10))
plt.scatter(cmip_nat_gdd_rmse[:-1],cmip_nat_gdd_sdr[:-1],c = "C3",s=80,zorder=2)
plt.scatter(nex_nat_gdd_rmse[:-1],nex_nat_gdd_sdr[:-1],c = "C2",s=80,zorder=2)

plt.scatter(cmip_nat_gdd_rmse[-1],cmip_nat_gdd_sdr[-1],c = "C3",s=80,marker="s",zorder=2)
plt.scatter(nex_nat_gdd_rmse[-1],nex_nat_gdd_sdr[-1],c = "C2",s=80,marker="s",zorder=2)
for i in range(int(len(cmip_nat_gdd_rmse))):
    ax.arrow(cmip_nat_gdd_rmse[i], cmip_nat_gdd_sdr[i], -cmip_nat_gdd_rmse[i]+nex_nat_gdd_rmse[i], -cmip_nat_gdd_sdr[i]+nex_nat_gdd_sdr[i],color="grey",zorder=1,alpha=0.6)
ax.set_xlabel("RMSE (days)")
ax.set_ylabel("STD ratio")
plt.show()
```

```python
# Plot GDD improvement
cmip_nat_gdd_rmse = cmip_rmse.groupby("AgVar").mean().loc["gdd"]
nex_nat_gdd_rmse = nex_rmse.groupby("AgVar").mean().loc["gdd"]

cmip_nat_gdd_sdr = cmip_sdr.groupby("AgVar").mean().loc["gdd"]
nex_nat_gdd_sdr = nex_sdr.groupby("AgVar").mean().loc["gdd"]

fig, ax = plt.subplots(1,1, figsize = (10,10))
plt.scatter(cmip_nat_gdd_rmse[:-1],cmip_nat_gdd_sdr[:-1],c = "C3",s=80,zorder=2)
plt.scatter(nex_nat_gdd_rmse[:-1],nex_nat_gdd_sdr[:-1],c = "C2",s=80,zorder=2)

plt.scatter(cmip_nat_gdd_rmse[-1],cmip_nat_gdd_sdr[-1],c = "C3",s=80,marker="s",zorder=2)
plt.scatter(nex_nat_gdd_rmse[-1],nex_nat_gdd_sdr[-1],c = "C2",s=80,marker="s",zorder=2)
for i in range(int(len(cmip_nat_gdd_rmse))):
    ax.arrow(cmip_nat_gdd_rmse[i], cmip_nat_gdd_sdr[i], -cmip_nat_gdd_rmse[i]+nex_nat_gdd_rmse[i], -cmip_nat_gdd_sdr[i]+nex_nat_gdd_sdr[i],color="grey",zorder=1,alpha=0.6)
ax.set_xlabel("RMSE (days)")
ax.set_ylabel("STD ratio")
plt.show()
```

```python
# Plot eGDD improvement
cmip_nat_egdd_rmse = cmip_rmse.groupby("AgVar").mean().loc["egdd"]
nex_nat_egdd_rmse = nex_rmse.groupby("AgVar").mean().loc["egdd"]

cmip_nat_egdd_sdr = cmip_sdr.groupby("AgVar").mean().loc["egdd"]
nex_nat_egdd_sdr = nex_sdr.groupby("AgVar").mean().loc["egdd"]

fig, ax = plt.subplots(1,1, figsize = (10,10))
plt.scatter(cmip_nat_egdd_rmse[:-1],cmip_nat_egdd_sdr[:-1],c = "C3",s=80,zorder=2)
plt.scatter(nex_nat_egdd_rmse[:-1],nex_nat_egdd_sdr[:-1],c = "C2",s=80,zorder=2)

plt.scatter(cmip_nat_egdd_rmse[-1],cmip_nat_egdd_sdr[-1],c = "C3",s=80,marker="s",zorder=2)
plt.scatter(nex_nat_egdd_rmse[-1],nex_nat_egdd_sdr[-1],c = "C2",s=80,marker="s",zorder=2)
for i in range(int(len(cmip_nat_egdd_rmse))):
    ax.arrow(cmip_nat_egdd_rmse[i], cmip_nat_egdd_sdr[i], -cmip_nat_egdd_rmse[i]+nex_nat_egdd_rmse[i], -cmip_nat_egdd_sdr[i]+nex_nat_egdd_sdr[i],color="grey",zorder=1,alpha=0.6)
ax.set_xlabel("RMSE (days)")
ax.set_ylabel("STD ratio")
plt.show()
```

```python
# Plot GDD improvement
cmip_nat_prcp_rmse = cmip_rmse.groupby("AgVar").mean().loc["prcp"]
nex_nat_prcp_rmse = nex_rmse.groupby("AgVar").mean().loc["prcp"]

cmip_nat_prcp_sdr = cmip_sdr.groupby("AgVar").mean().loc["prcp"]
nex_nat_prcp_sdr = nex_sdr.groupby("AgVar").mean().loc["prcp"]

fig, ax = plt.subplots(1,1, figsize = (10,10))
plt.scatter(cmip_nat_prcp_rmse[:-1],cmip_nat_prcp_sdr[:-1],c = "C3",s=80,zorder=2)
plt.scatter(nex_nat_prcp_rmse[:-1],nex_nat_prcp_sdr[:-1],c = "C2",s=80,zorder=2)

plt.scatter(cmip_nat_prcp_rmse[-1],cmip_nat_prcp_sdr[-1],c = "C3",s=80,marker="s",zorder=2)
plt.scatter(nex_nat_prcp_rmse[-1],nex_nat_prcp_sdr[-1],c = "C2",s=80,marker="s",zorder=2)
for i in range(int(len(cmip_nat_prcp_rmse))):
    ax.arrow(cmip_nat_prcp_rmse[i], cmip_nat_prcp_sdr[i], -cmip_nat_prcp_rmse[i]+nex_nat_prcp_rmse[i], -cmip_nat_prcp_sdr[i]+nex_nat_prcp_sdr[i],color="grey",zorder=1,alpha=0.6,head_width=False,width=0.0001)
ax.set_xlabel("RMSE (m m$^{-2}$)")
ax.set_ylabel("STD ratio")
plt.show()
```

```python
# Plot all
fig, ax = plt.subplots(1,3, figsize = (18,10), sharey = True)

#GDD
ax[0].scatter(cmip_nat_gdd_rmse[:-1],cmip_nat_gdd_sdr[:-1],c = "C3",s=80,zorder=2)
ax[0].scatter(nex_nat_gdd_rmse[:-1],nex_nat_gdd_sdr[:-1],c = "C2",s=80,zorder=2)

ax[0].scatter(cmip_nat_gdd_rmse[-1],cmip_nat_gdd_sdr[-1],c = "C3",s=80,marker="s",zorder=2)
ax[0].scatter(nex_nat_gdd_rmse[-1],nex_nat_gdd_sdr[-1],c = "C2",s=80,marker="s",zorder=2)
for i in range(int(len(cmip_nat_gdd_rmse))):
    ax[0].arrow(cmip_nat_gdd_rmse[i], cmip_nat_gdd_sdr[i], -cmip_nat_gdd_rmse[i]+nex_nat_gdd_rmse[i], -cmip_nat_gdd_sdr[i]+nex_nat_gdd_sdr[i],color="grey",zorder=1,alpha=0.6)

# eGDD
ax[1].scatter(cmip_nat_egdd_rmse[:-1],cmip_nat_egdd_sdr[:-1],c = "C3",s=80,zorder=2)
ax[1].scatter(nex_nat_egdd_rmse[:-1],nex_nat_egdd_sdr[:-1],c = "C2",s=80,zorder=2)

ax[1].scatter(cmip_nat_egdd_rmse[-1],cmip_nat_egdd_sdr[-1],c = "C3",s=80,marker="s",zorder=2)
ax[1].scatter(nex_nat_egdd_rmse[-1],nex_nat_egdd_sdr[-1],c = "C2",s=80,marker="s",zorder=2)
for i in range(int(len(cmip_nat_egdd_rmse))):
    ax[1].arrow(cmip_nat_egdd_rmse[i], cmip_nat_egdd_sdr[i], -cmip_nat_egdd_rmse[i]+nex_nat_egdd_rmse[i], -cmip_nat_egdd_sdr[i]+nex_nat_egdd_sdr[i],color="grey",zorder=1,alpha=0.6)

# Precip
ax[2].scatter(cmip_nat_prcp_rmse[:-1],cmip_nat_prcp_sdr[:-1],c = "C3",s=80,zorder=2)
ax[2].scatter(nex_nat_prcp_rmse[:-1],nex_nat_prcp_sdr[:-1],c = "C2",s=80,zorder=2)

ax[2].scatter(cmip_nat_prcp_rmse[-1],cmip_nat_prcp_sdr[-1],c = "C3",s=80,marker="s",zorder=2)
ax[2].scatter(nex_nat_prcp_rmse[-1],nex_nat_prcp_sdr[-1],c = "C2",s=80,marker="s",zorder=2)
for i in range(int(len(cmip_nat_prcp_rmse))):
    ax[2].arrow(cmip_nat_prcp_rmse[i], cmip_nat_prcp_sdr[i], -cmip_nat_prcp_rmse[i]+nex_nat_prcp_rmse[i], -cmip_nat_prcp_sdr[i]+nex_nat_prcp_sdr[i],color="grey",zorder=1,alpha=0.6,head_width=False,width=0.0001)

# Labels
ax[0].scatter(0.,1.0,c = "C3",s=80, marker = "s", label = "CMIP ensemble mean")
ax[0].scatter(0.,1.0,c = "C2",s=80, marker = "s", label = "NEX ensemble mean")

ax[0].scatter(0.,1.0,c = "C3",s=80, label = "CMIP models")
ax[0].scatter(0.,1.0,c = "C2",s=80, label = "NEX models")

ax[0].scatter(0.,1.0,c = "C4",s=120, marker = "D", label = "'ideal' model")

ax[1].scatter(0.,1.0,c = "C4",s=120, marker = "D")
ax[2].scatter(0.,1.0,c = "C4",s=120, marker = "D")
    
ax[0].set_xlabel("RMSE (days)")
ax[0].set_title("GDD")
ax[1].set_xlabel("RMSE (days)")
ax[1].set_title("eGDD")
ax[2].set_xlabel("RMSE ($m\cdot m^{-2}$)")
ax[2].set_title("Precip")

ax[0].set_ylabel("SD ratio")
plt.tight_layout()
ax[0].legend()
# plt.savefig("./output/figs/improvements.png",dpi=500)
plt.show()
```

### Yield

```python
# Read NEX
nex_rmse = pd.read_csv("../county/output/data/nex_yield_rmse_56-05.csv")
nex_rmse["GEOID"] = nex_rmse["GEOID"].astype(str).str.zfill(5)
nex_rmse.set_index(["GEOID"], inplace = True)

nex_sdr = pd.read_csv("../county/output/data/nex_yield_sdratio_56-05.csv")
nex_sdr["GEOID"] = nex_sdr["GEOID"].astype(str).str.zfill(5)
nex_sdr.set_index(["GEOID"], inplace = True)
```

```python
# Read CMIP
cmip_rmse = pd.read_csv("../county/output/data/cmip_yield_rmse_56-05.csv")
cmip_rmse["GEOID"] = cmip_rmse["GEOID"].astype(str).str.zfill(5)
cmip_rmse.set_index(["GEOID"], inplace = True)

cmip_sdr = pd.read_csv("../county/output/data/cmip_yield_sdratio_56-05.csv")
cmip_sdr["GEOID"] = cmip_sdr["GEOID"].astype(str).str.zfill(5)
cmip_sdr.set_index(["GEOID"], inplace = True)
```

```python
# Plot yield improvement from ARITHMETIC MEAN
cmip_nat_rmse = cmip_rmse.mean()
nex_nat_rmse = nex_rmse.mean()

cmip_nat_sdr = cmip_sdr.mean()
nex_nat_sdr = nex_sdr.mean()

fig, ax = plt.subplots(1,1, figsize = (10,10))

plt.scatter(cmip_nat_rmse[:-2],cmip_nat_sdr[:-2],c = "C3",s=80,zorder=2)
plt.scatter(nex_nat_rmse[:-2],nex_nat_sdr[:-2],c = "C2",s=80,zorder=2)

plt.scatter(cmip_nat_rmse[-2],cmip_nat_sdr[-2],c = "C3",s=80,marker="s",zorder=2)
plt.scatter(nex_nat_rmse[-2],nex_nat_sdr[-2],c = "C2",s=80,marker="s",zorder=2)

for i in range(int(len(cmip_nat_prcp_rmse))):
    ax.arrow(cmip_nat_rmse[i], cmip_nat_sdr[i], -cmip_nat_rmse[i]+nex_nat_rmse[i], -cmip_nat_sdr[i]+nex_nat_sdr[i],color="grey",zorder=1,alpha=0.6,head_width=False,width=0.0001)
    
ax.scatter(0.,1.0,c = "C3",s=80, marker = "s", label = "CMIP ensemble mean")
ax.scatter(0.,1.0,c = "C2",s=80, marker = "s", label = "NEX ensemble mean")

ax.scatter(0.,1.0,c = "C3",s=80, label = "CMIP models")
ax.scatter(0.,1.0,c = "C2",s=80, label = "NEX models")

ax.scatter(cmip_nat_rmse[-1],cmip_nat_sdr[-1],c = "C4",s=80, marker = "D", label = "GMFD")

ax.scatter(0.,1.0,c = "C8",s=150, marker = "h", label = "'ideal' model")

ax.set_xlabel("RMSE")
ax.set_ylabel("SD ratio")

plt.tight_layout()
ax.legend()

# plt.savefig("./output/figs/yield_improvement.png", dpi = 500)
plt.show()
```

```python
################ Plot yield improvement from ARITHMETIC MEAN

# USDA weights
usda_weights = pd.read_csv("./weights/usda_county_to_national_weights.csv")
usda_weights["GEOID"] = usda_weights["GEOID"].astype(str).str.zfill(5)
usda_weights = usda_weights.groupby('GEOID').mean()
```

```python
# Get weighted means
cmip_nat_rmse = pd.merge(usda_weights.filter(['GEOID','national_weight']), cmip_rmse.reset_index(), on = "GEOID", how = "inner")
for col in cmip_nat_rmse.columns[2:]:
    cmip_nat_rmse[col] = cmip_nat_rmse[col] * cmip_nat_rmse['national_weight']
cmip_nat_rmse = cmip_nat_rmse.sum().drop(['GEOID', 'national_weight'])

nex_nat_rmse = pd.merge(usda_weights.filter(['GEOID','national_weight']), nex_rmse.reset_index(), on = "GEOID", how = "inner")
for col in nex_nat_rmse.columns[2:]:
    nex_nat_rmse[col] = nex_nat_rmse[col] * nex_nat_rmse['national_weight']
nex_nat_rmse = nex_nat_rmse.sum().drop(['GEOID', 'national_weight'])

cmip_nat_sdr = pd.merge(usda_weights.filter(['GEOID','national_weight']), cmip_sdr.reset_index(), on = "GEOID", how = "inner")
for col in cmip_nat_sdr.columns[2:]:
    cmip_nat_sdr[col] = cmip_nat_sdr[col] * cmip_nat_sdr['national_weight']
cmip_nat_sdr = cmip_nat_sdr.sum().drop(['GEOID', 'national_weight'])

nex_nat_sdr = pd.merge(usda_weights.filter(['GEOID','national_weight']), nex_sdr.reset_index(), on = "GEOID", how = "inner")
for col in nex_nat_sdr.columns[2:]:
    nex_nat_sdr[col] = nex_nat_sdr[col] * nex_nat_sdr['national_weight']
nex_nat_sdr = nex_nat_sdr.sum().drop(['GEOID', 'national_weight'])
```

```python
# Plot yield improvement from WEIGHTED MEAN
fig, ax = plt.subplots(1,1, figsize = (10,10))

plt.scatter(cmip_nat_rmse[:-2],cmip_nat_sdr[:-2],c = "C3",s=80,zorder=2)
plt.scatter(nex_nat_rmse[:-2],nex_nat_sdr[:-2],c = "C2",s=80,zorder=2)

plt.scatter(cmip_nat_rmse[-2],cmip_nat_sdr[-2],c = "C3",s=80,marker="s",zorder=2)
plt.scatter(nex_nat_rmse[-2],nex_nat_sdr[-2],c = "C2",s=80,marker="s",zorder=2)

for i in range(int(len(cmip_nat_prcp_rmse))):
    ax.arrow(cmip_nat_rmse[i], cmip_nat_sdr[i], -cmip_nat_rmse[i]+nex_nat_rmse[i], -cmip_nat_sdr[i]+nex_nat_sdr[i],color="grey",zorder=1,alpha=0.6,head_width=False,width=0.0001)
    
ax.scatter(0.,1.0,c = "C3",s=80, marker = "s", label = "CMIP ensemble mean")
ax.scatter(0.,1.0,c = "C2",s=80, marker = "s", label = "NEX ensemble mean")

ax.scatter(0.,1.0,c = "C3",s=80, label = "CMIP models")
ax.scatter(0.,1.0,c = "C2",s=80, label = "NEX models")

ax.scatter(cmip_nat_rmse[-1],cmip_nat_sdr[-1],c = "C4",s=80, marker = "D", label = "GMFD")

ax.scatter(0.,1.0,c = "C8",s=150, marker = "h", label = "'ideal' model")

ax.set_xlabel("RMSE")
ax.set_ylabel("SD ratio")

plt.tight_layout()
ax.legend()

# plt.savefig("./output/figs/yield_improvement.png", dpi = 500)
plt.show()
```
