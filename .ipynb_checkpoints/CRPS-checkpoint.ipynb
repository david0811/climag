```python
import pandas as pd
import numpy as np
import scipy.stats as ss
import scipy.integrate as integrate
from statsmodels.distributions.empirical_distribution import ECDF
import matplotlib.pyplot as plt

import seaborn as sns; sns.set()
```

```python
nex = pd.read_csv('./analysis/combine/output/nex_agvar_hist.csv')
cmip = pd.read_csv('./analysis/combine/output/cmip_agvar_all.csv')
```

```python
data = nex.copy().dropna()
data["GEOID"] = data["GEOID"].astype(str).str.zfill(5)
```

```python
def make_sa(data):
    return (data - data.mean())/np.std(data)
```

```python
res = []
nx = 10000
for agvar in data['AgVar'].unique():
    nnex = data.query('Year >= 1956 and Year <= 2005 and AgVar == "' + agvar + '"')
    for geoid in nnex['GEOID'].unique():
        ens = make_sa(nnex.query('GEOID == "' + geoid + '"').drop(columns = ['GEOID', 'GMFD', 'ensemble_mean', 'Year', 'AgVar']).values.flatten())
        obs = make_sa(nnex.query('GEOID == "' + geoid + '"')['GMFD'].values)
        
        ens_kde = ss.gaussian_kde(ens)
        obs_kde = ss.gaussian_kde(obs)

        ens_ecdf = ECDF(ens)
        obs_ecdf = ECDF(obs)
        
        cprs = integrate.trapz(y = (ens_ecdf(np.linspace(0.5*ens.min(),2*ens.max(),nx))-obs_ecdf(np.linspace(0.5*ens.min(),2*ens.max(),nx)))**2, x = np.linspace(0.5*ens.min(),2*ens.max(),nx))
        
        res.append({'AgVar' : agvar, 'GEOID' : geoid, 'CPRS' : cprs})
```

```python
datares = pd.DataFrame(res)
```

```python
datares.plot()
```

```python
kwargs = {'cumulative': True}
sns.distplot(ens, hist_kws=kwargs, kde_kws=kwargs)
sns.distplot(obs, hist_kws=kwargs, kde_kws=kwargs)
plt.show()
```
