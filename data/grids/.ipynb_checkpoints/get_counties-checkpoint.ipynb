### A simple script to get only the conterminous counties/states of the USA

```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gp
from shapely.geometry import *
import numpy as np
```

```python
# Read in county data
counties = gp.read_file("../../conus_shp/tl_2018_us_county/tl_2018_us_county.shp")
```

```python
counties.head()
```

```python
del counties["COUNTYNS"]
del counties["NAMELSAD"]
del counties["LSAD"]
del counties["CLASSFP"]
del counties["MTFCC"]
del counties["CSAFP"]
del counties["CBSAFP"]
del counties["METDIVFP"]
del counties["FUNCSTAT"]
del counties["ALAND"]
del counties["AWATER"]
del counties["INTPTLAT"]
del counties["INTPTLON"]
```

```python
counties.head()
```

```python
counties["STATEFP"].unique()
```

```python
counties = counties.query("STATEFP != '02' and STATEFP != '15' and STATEFP != '60' and STATEFP != '66' and STATEFP != '72' and STATEFP != '78' and STATEFP != '69'")
```

```python
counties.plot(figsize=(50,20))
```

```python
# Save
counties.to_file("./contig_county/counties_contig.shp")
```
