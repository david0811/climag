##Â Area fractions: how much land per pixel was harvested for Corn?

```python
# http://www.earthstat.org/harvested-area-yield-4-crops-1995-2005/
# http://mapspam.info
maize_95 = rio.open("./earthstat/HarvAreaYield_4Crops_95-00-05_Geotiff/Maize/Maize_1995_Area.tif")
maize_00 = rio.open("./earthstat/HarvAreaYield_4Crops_95-00-05_Geotiff/Maize/Maize_2000_Area.tif")
maize_05 = rio.open("./earthstat/HarvAreaYield_4Crops_95-00-05_Geotiff/Maize/Maize_2005_Area.tif")
```

```python
print(maize_95.meta)
print(maize_00.meta)
print(maize_05.meta)
```

```python
# conterminous USA
crop_extent = gp.read_file("./usa_bounds/usa-boundary-dissolved.shp")
```

```python
print(crop_extent.crs)
```

```python
# create geojson object from the shapefile imported above
extent_geojson = mapping(crop_extent['geometry'][0])
```

```python
# Crop
maize_95_crop, maize_crop_affine = mask(maize_95, [extent_geojson], crop=True)

# Create spatial plotting extent for the cropped layer
maize_extent = plotting_extent(maize_95_crop[0], maize_crop_affine)
```

```python
# Take a look
fig, ax = plt.subplots(figsize=(12, 10))
m_plot = ax.imshow(maize_95_crop[0],
          extent=maize_extent, cmap="Greys", vmin = -0.1, vmax = 1.0)
fig.colorbar(m_plot, fraction=.023, ax=ax)
ax.set_title("1995 Harvested Corn Area")
ax.set_axis_off()
```

##Â Convert to GeoPandas DataFrame

```python
# Transform data for GeoPandas
T0 = maize_crop_affine
```

```python
maize_crop_affine
```

```python
# https://rasterio.readthedocs.io/en/stable/topics/georeferencing.html?highlight=affine
T_top_right = T0 * Affine.translation(1.0, 0.0)
T_bottom_right = T0 * Affine.translation(1.0, 1.0)
T_bottom_left = T0 * Affine.translation(0.0, 1.0)
```

```python
# Transformation function for each corner
xy_tl = lambda r, c: (c, r) * T0
xy_tr = lambda r, c: (c, r) * T_top_right
xy_br = lambda r, c: (c, r) * T_bottom_right
xy_bl = lambda r, c: (c, r) * T_bottom_left
```

```python
# Lats and lons of each pixel
raster_geo = gp.GeoSeries([Polygon([xy_tl(j,i),xy_tr(j,i),xy_br(j,i),xy_bl(j,i)]) for i in range(maize_95_crop[0][0].size) for j in range(maize_95_crop[0][:,1].size)])
```

```python
# Fill arrays
raster_95_values = [maize_95_crop[0][j,i] if maize_95_crop[0][j,i] >= 0.0 else 0.0 for i in range(maize_95_crop[0][0].size) for j in range(maize_95_crop[0][:,1].size)]
raster_00_values = [maize_00_crop[0][j,i] if maize_00_crop[0][j,i] >= 0.0 else 0.0 for i in range(maize_00_crop[0][0].size) for j in range(maize_00_crop[0][:,1].size)]
raster_05_values = [maize_05_crop[0][j,i] if maize_05_crop[0][j,i] >= 0.0 else 0.0 for i in range(maize_05_crop[0][0].size) for j in range(maize_05_crop[0][:,1].size)]
```

```python
# Final GeoDataFrame
raster_geoP = gp.GeoDataFrame({"geometry": raster_geo, "frac_95": raster_95_values, "frac_00": raster_00_values, "frac_05": raster_05_values})
```

```python
# Take average
raster_geoP["frac"] = (raster_geoP["frac_95"] + raster_geoP["frac_00"] + raster_geoP["frac_05"]) / 3.0

del raster_geoP["frac_95"]
del raster_geoP["frac_00"]
del raster_geoP["frac_05"]
```

```python
# Convert to same projection as county data
raster_geoP.crs = {'init' :'epsg:4326'}
raster_geoP = raster_geoP.to_crs({'init': 'epsg:4269'})
```

```python
# Save
raster_geoP.to_file("./output/area_fracs.shp")
```
